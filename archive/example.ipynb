```python
import subprocess
from llama_index.core import Settings
from llama_index.llms.openai_like import OpenAILike
import requests
import time
```

```python
def run_embed_model_on_gpu():
    result = subprocess.run(["docker", "compose", "-f", "docker-compose.embed.yaml", "up", "-d"])
    if result.returncode == 0:
        print("Embed сервис запускается на gpu, ожидайте ...")
        start = time.time()
        while True:
            try:
                time.sleep(1)
                req = requests.get(url="http://localhost:8000/health")
                print(req.status_code)
                print("✅ vLLM сервис жив здоров")
                Settings.llm = OpenAILike(
                model="qwen3-embed",                    
                api_base="http://localhost:8000/v1", 
                )
                print("Embed модель подключена")
                break
            except Exception as e:
                end = time.time()
            if end - start > 120:
                print(f"Ошибка, сервис не может запуститься {e}")
                break
    else:
        print(f"Ошибка: {result.stderr}")
```

```python
run_embed_model_on_gpu()
```
